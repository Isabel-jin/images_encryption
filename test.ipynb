```python
img = cv2.imread('D:/lena.png')
cv2.rectangle(img,(10,10),(150,150),(0,255,0),2)
x=img[10:150,10:150]
plt.imshow(x)
```

```python
x0 = x[:, :, 0]
x1 = x[:, :, 1]
x2 = x[:, :, 2]
r = np.array(np.zeros(8))
print(r.dtype)
r = np.array([0.343, 0.432, 0.63, 3.769, 3.82, 3.8, 0.1, 1])
e0 = jiami(x0, r)
e1 = jiami(x1, r)
e2 = jiami(x2, r)
e = np.dstack((e0, e1, e2))
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(e, cv2.COLOR_BGR2RGB))
```
